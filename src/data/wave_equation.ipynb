```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def fn_u(t,x):
    return 1/2*np.sin(np.pi*x)*np.cos(np.pi*t) + 1/3*np.sin(3*np.pi*x)*np.sin(3*np.pi*t)
```

```python
def fn_v(t,x):
    #return the time derivative of the wavefunction
    return -1/2*np.pi*np.sin(np.pi*x)*np.sin(np.pi*t) + 1/3*3*np.pi*np.sin(3*np.pi*x)*np.cos(3*np.pi*t)
```

```python
# create the x and t values
n_point = 31
x = np.linspace(0,1,n_point)
t = np.linspace(0,2,2000)
```

```python
# get the meshgrid
X,T = np.meshgrid(x,t)

# get the u and v values
U = fn_u(T,X)
V = fn_v(T,X)
```

```python
# plot U at t = 0, 380, 760, 1140
fig, ax = plt.subplots(2,4, figsize=(10,4))
fig.suptitle("Wace equation at t = 0, 380, 760, 1140")
ax = ax.flatten()
for i, t_val in enumerate([0, 380, 760, 1140]):
    ax[i].plot(x, U[t_val], label="U")
    ax[i+4].plot(x, V[t_val], label="V")
    ax[i].set_title(f"t = {t_val/1000}")
    ax[i+4].set_xlabel("t")
    ax[i].set_ylim(-1,1)
    ax[i+4].set_ylim(-5,5)
# only leaves the y label on the leftmost plot
for i in range(1,4):
    ax[i].set_yticklabels([])
    ax[i+4].set_yticklabels([])
ax[0].set_ylabel("U")
ax[4].set_ylabel("V")
plt.tight_layout()
plt.savefig(("../../result/wave_equation.png"))
plt.show()
```

```python
# plot V at t = 0, 380, 760, 1140
fig, ax = plt.subplots(1,4, figsize=(10,3))
fig.suptitle("v at t = 0, 380, 760, 1140")
ax = ax.flatten()

for i, t_val in enumerate([0, 380, 760, 1140]):
    ax[i].plot(x, V[t_val], label="V")
    ax[i].set_title(f"t = {t_val/1000}")
    ax[i].set_xlabel("t")
    ax[i].set_ylim(-5,5)
plt.tight_layout()
plt.savefig(("../../result/wave_equation_v.png"))
plt.show()

```
